```python
"""
用for循环，索引,冒泡循环
"""
b = [4,3,2,1]
length = len(b)
for i in range(length - 1):
    for j in range(length - i - 1):#- 1 是锁定最大值
        if b[j] > b[j + 1]:
            b[j],b[j + 1] = b[j + 1],b[j]
print(b)
```

## 测试字符串

> - isalnum() 中是不能包含空格，否则会返回False

```python
"""
给一个密码，长度是8位，含有大小写，含有数字，如果不满足则显示密码必须含有大写.....
四个判断条件
"""
password = input(":>>")
N1 = 0
N2 = 0
N3 = 0

if len(password) <= 8:
    print("密码必须大于8位")
for i in password:
    if i.isupper():
        N1 += 1
    if i.islower():
        N2 += 1
    if i.isdigit():
        N3 += 1
if N1 == 0:
    print("密码必须有大写")
if N2 == 0:
    print("密码必须有小写")
if N3 == 0:
    print("密码必须有字母")
```

```python
import re
"""
re是python里面的库
"""
def check(str):
    """
    用re库里的complie()这个方法去编译【r'\d{3}-\d{2}-\d{4}'】这个正则表达式
    """
    result = re.compile(r'\d{3}-\d{2}-\d{4}')
    if result.match(str):
        """
        match是根据reasult这个规则去匹配字符串
        """
        print("Valid SSN")
    else:
        print("Invalid SSN")
```

```python
"""
给一个密码，长度是8个字符，包含英文和数字，至少含有两个数字
三个判断条件
"""
password = input(":>>")
N2 = 0
N3 = 0

if len(password) <= 8:
    print("invalid password")
    for i in password:
        if i.isalnum():
            N2 += 1
        if i.isdigit():
            N3 += 1
    if N2 == 0:
        print("invalid password")
    if N3 == 0:
        print("invalid password")
else:
    print("valid password")
```
